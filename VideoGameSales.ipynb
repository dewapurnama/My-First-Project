Saya disini akan mengambil contoh variabel yaitu Year.

Untuk Year sendiri mempunyai mean 2006 dan median 2007. Ini berarti bahwa untuk variabel Year, pusat distribusi atau penyebaran data nya berkisar di Tahun 2006 dan 2007.

IQR = Q3-Q1 = 2010-2003 = 7

1,5 x 7 = 1,5 x IQR = 1,5 x 7 = 10,5

Q1 - (1,5 x IQR) = 2003 - 7 = 1994

Q3 + (1,5 X 1QR) = 2010 + 7 = 2017

Maka, bisa ditentukan bahwa outlier merupakan data yang berada di bawah 1994 atau di atas 2017

Analisis :
    
Box plot digunakan untuk menjelaskan karakteristik dari sebuah variabel kuantitatif. Box plot diatas merupakan box plot tanpa outlier. Box plot sendiri dibedakan menjadi tiga bagian utama yaitu whisker bawah, box dan whisker atas. Whisker bawah merupakan nilai minimum setelah outlier dikeluarkan yang didapat dari Q1 - (1,5 x IQR). Box sendiri terdiri dari tiga garis yaitu garis bawah sebagai kuartil bawah, garis tengah sebagai median, dan garis paling atas sebagai kuartil atas. Adapun whisker atas merupakan nnilai maksimum setelah outlier dikeluarkan yang didapat dari Q3 + (1,5 x IQR). Median berada pada angka 0,01, kuartil atas ada pada angka  0,04, dan whisker atas memiliki rentang dari 0,04 sampai dengan 0,10. Angka yang terletak di bawah whisker bawah atau yang terletak di atas whisker atas akan dianggap sebagai outliers.

```python
plt.scatter(df['Global_Sales'], df['Other_Sales'])
plt.show()
```

Analisis :

Scatterplot digunakan untuk menunjukkan hubungan diantara dua variabel kuantitatif. Disini saya menggunakan variabel Global_Sales dan Other_Sales.

Adapun pola keseluruhannya adalah sebagai berikut :

Form : Bentuk dari data di atas yaitu mendekati/hampir linear.

Direction : Karena kenaikan satu variabel diikuti dengan kenaikan variabel yang lain, maka arah dari scatterplot tersebut adalah positif.

Strength : Kekuatan dari suatu scatterplot adalah seberapa dekat data yang ada mengikuti bentuk dari scatterplot itu sendiri. Dapat dilihat dari scatterplot di atas bahwa sebagian data mengikuti bentuknya, namun sebagian lagi tidak, sehingga kekuatan dari scatterplot ini adalah sedang.

Setelah menentukan pola keseluruhan, maka dapa ditentukan pula outliernya. Outlier pada scatterplot sendiri merupakan data yang tidak mengikuti form/bentuk scatterplot itu sendiri. Dapat dilihat dari scatterplot di atas, ada cukup banyak outlier yang menyebar diluar bentuk atau pola scatterplotnya.



# CHAPTER 2.3, 2.4, 2.5

```python
x = df["Global_Sales"]
y = df["Other_Sales"]
slope, intercept, r, p, std_err = stats.linregress(x,y) 
```

```python
print('slope={0}, intercept={1}, r={2}, p={3}, std_err ={4} '.format(slope, intercept, r,p,std_err))
```

```python
def myfunc(x):
  return slope * x + intercept
```

```python
mymodel = list(map(myfunc, x)) 
```

```python
plt.scatter(x, y) 
plt.plot(x, mymodel)
```

Analisis:
    
Scatter plot diatas membandingkan dua variabel kuantitatif yaitu Global_Sales dan Other_Sales. Scatter plot digunakan untuk menunjukkan korelasi dan kekuatan antar dua variabel itu sendiri. Pada scatter plot di atas, ditemukan bahwa arahnya yaitu positif (variable Global_Sales dan Other_Sales bertolak belakang). Adapun kekuatan scatter plot tersebut yaitu sedang yang bisa dilihat dari korelasi(r) yang besar yaitu berkisar 0,7.

```python
pd.crosstab(df.Platform, df.Genre)
```

Analisis:

Two way table diatas menunjukkan hubungan antara 2 variabel kategorikal, dalam hal ini Platform dan Genre. Misal, pada tabel di atas bisa dilihat bahwa ada 61 video games yang terjual pada platform 2600 dengan Genre Action.

```python
pd.crosstab(df.Platform, df.Genre, margins=True, margins_name="Total")
```

Analisis:

Two way table diatas menunjukkan hubungan antara 2 variabel kategorikal, dalam hal ini Platform dan Genre namun dengan memasukkan total untuk setiap data pada masing-masing variabel. Bisa dilihat total data dari masing-masing variabel adalah 16598 (sesuai dengan banyaknya jumlah data pada data set).

```python
pd.crosstab(df.Platform, df.Genre, normalize=True, margins=True, margins_name="Total")
```

Analisis:

Two way table di atas menunjukkan perbandingan populasi antar dua variabel kategorikal dengan kesuluruhan data. Misal, terdapat 0,003 video games genre action yang dijual di platform 2600 dari total keseluruhan data yang berjumlah 1.
